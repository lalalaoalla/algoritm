```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        k = n/2
        counts = {}
        res = 0
        max_value = 0
        for i in nums:
            if i in counts:
                counts[i]+=1
            else:
                counts[i] = 1 
        for key, value in counts.items():
            if value>= max_value:
                max_value = value
                res = key
                if max_value >= k:#литкод не пропускает со знаком равно
                    return res


nums = [3,2,3]

solution = Solution()
res = solution.majorityElement(nums)

print(res)
        
```

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        n = len(nums)
        for num in set(nums):  # iterate over unique elements in nums
            if nums.count(num) > n // 2:
                return num
            
nums = [3,2,3]

solution = Solution()
res = solution.majorityElement(nums)

print(res)
```
