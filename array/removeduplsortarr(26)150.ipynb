Собственно решается это задание примерно также, как и задача 27, только почему-то еси идти с начала списка, то что-то идет не так

Если идти с начала, то что-то идет не так со сдвигами

```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        k = len(nums)
        if len(nums) == None:
            return None
        if len(nums) == 1:
            return 1
        for i in range(len(nums)-1,-1,-1):
            if nums[i] == nums[i-1]:
                nums[i] = '_'
                nums.append(nums.pop(i))
                k-=1
        return k
    
nums = [0,0,1,1,1,2,2,3,3,4]

solution = Solution()

res = solution.removeDuplicates(nums)
print(res)
```
