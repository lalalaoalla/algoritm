```python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        n1 = n-1
        m1 = m-1
        nm=n+m -1
        while n1 >= 0 and m1 >= 0:
            if nums2[n1] >= nums1[m1]:
                nums1[nm] = nums2[n1]
                n1 -= 1
            else:
                nums1[nm] = nums1[m1]
                m1 -= 1
            nm -= 1
        while n1 >= 0:
            nums1[nm] = nums2[n1]
            n1 -= 1
            nm -= 1

nums1 = [1,3,5,0,0]
nums2 = [2,6]

n = 2
m = 3

solution = Solution()
solution.merge(nums1, m, nums2, n)

print(nums1)
```
