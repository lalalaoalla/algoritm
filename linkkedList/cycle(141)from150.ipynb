Использем быстрые и медленные указатели, если вдруг они оказались равны, то цикл есть

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        slow = fast = head
        boolr = False
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if fast == slow:
                boolr = True
                return boolr
        return boolr
        
```
