```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def getDecimalValue(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: int
        """
        node1 = head
        prev = None
        while node1:
            next_node1 = node1.next
            node1.next = prev
            prev = node1
            node1 = next_node1
        value = 0
        st = 0
        while prev:
            #print(prev.val)
            value += prev.val*2**st
            print (f' prev {prev.val},   value {value}')
            st+=1
            prev=prev.next
        return value
    
head = ListNode(1,ListNode(0, ListNode(1)))#  - 10

solution = Solution()
res = solution.getDecimalValue(head)

print(res)
        
```
