```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def isPalindrome(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: bool
        """
        # 1. Преобразовать связанный список в список Python
        values = []
        node = head
        while node:
            values.append(node.val)
            node = node.next

        # 2. Проверить, является ли список Python палиндромом
        return values == values[::-1]  # Сравнение списка с его обратной копией

# Пример использования:
head = ListNode(1, ListNode(2, ListNode(2, ListNode(1))))
#head = ListNode(1, ListNode(2))
solution = Solution()
res = solution.isPalindrome(head)
print(res)  # Вывод: True

head2 = ListNode(1, ListNode(2, ListNode(3, ListNode(2, ListNode(1)))))

solution = Solution()
res2 = solution.isPalindrome(head2)
print(res2) # True

head3 = ListNode(1, ListNode(2, ListNode(3)))
solution = Solution()
res3 = solution.isPalindrome(head3)
print(res3) # False
```

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def isPalindrome(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: bool
        """
        # 1. Найти середину списка (slow/fast pointers)
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # 2. Развернуть вторую половину списка
        prev = None
        curr = slow
        while curr:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node

        # 3. Сравнить первую и развернутую вторую половины
        left = head
        right = prev  # prev теперь указывает на начало развернутой второй половины
        while right:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next

        return True

# Пример использования:
head = ListNode(1, ListNode(2, ListNode(2, ListNode(1))))

solution = Solution()

res = solution.isPalindrome(head)
print(res)
```
