```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def isPalindrome(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: bool
        """
        # 1.    (slow/fast pointers)
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # 2.    
        prev = None
        curr = slow
        while curr:
            next_node = curr.next
            curr.next = prev
            prev = curr
            curr = next_node

        # 3.      
        left = head
        right = prev  # prev       
        while right:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next

        return True

#  :
head = ListNode(1, ListNode(2, ListNode(2, ListNode(1))))

solution = Solution()

res = solution.isPalindrome(head)
print(res)
```
