```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: Optional[ListNode]
        :type l2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        node1 = l1
        node2 = l2
        res = ListNode()
        carry = 0
        value = 0
        
        value = node1.val + node2.val + carry
        carry = value // 10
        res.val = value % 10
        head = res
        node1 = node1.next
        node2 = node2.next
        while node1 or node2:
            value = carry
            if node1 != None:
                value += node1.val
                node1 = node1.next
            if node2 != None:
                value+= node2.val
                node2= node2.next
            carry = value // 10
            res.next = ListNode(value%10)
            res = res.next
        if carry!=0:
            res.next = ListNode(carry)
        return head
                

list1 = ListNode(1,ListNode(2,ListNode(2)))
list2 = ListNode(4, ListNode(5, ListNode(9)))

solution = Solution()

res = solution.addTwoNumbers(list1, list2)

while res:
    print(res.val)
    res = res.next
```
