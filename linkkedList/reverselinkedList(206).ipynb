```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head):#: Optional[ListNode]) -> Optional[ListNode]:
        if (head is None) or (head.next is None):
            return head
        i = head
        prev = None#ссылка на предыдущий элемент
        while i: 
            next_node = i.next#next_node - ссылка на след узел
            i.next = prev# теперь ссылка идет на предыдущий элемент
            prev = i# переместили prev на текущий узел
            i = next_node# а теперь i равен следующему
        return prev

head = ListNode(1,ListNode(2, ListNode(3,ListNode(4, ListNode(5)))))

solution = Solution()
res = solution.reverseList(head)
while res:
    print(res.val)
    res = res.next
```
