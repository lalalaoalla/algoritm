```python
#    -!!!!!!!!!!!!!11

# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        node1 = list1
        node2 = list2
        res = ListNode()
        
        if node1 == None and node2 == None:
            return None
        if node1 == None and node2 != None:
            return node2
        if node1 != None and node2 == None:
            return node1

        if node1.val <= node2.val:
            res = node1
            node1 = node1.next
        else:
            res = node2
            node2 = node2.next
        head = res

        while node1 and node2:
            if node1.val <= node2.val:
                res.next = node1
                node1 = node1.next
                res = res.next
            else:
                res.next = node2
                node2 = node2.next
                res = res.next
        if node1 == None and node2 != None:
            res.next = node2
            return head
        if node1 != None and node2 == None:
            res.next = node1
            return head
        return head
        
list1 = ListNode(1,ListNode(2,ListNode(2)))
list2 = ListNode(4, ListNode(5, ListNode(6)))

solution = Solution()

res = solution.mergeTwoLists(list1, list2)

while res:
    print(res.val)
    res = res.next
```
